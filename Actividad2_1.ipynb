**2.- Escriba una función que reciba como parámetro un número entero 
positivo y entregue como resultado todos los divisores de ese número.**

```python
# Crea función 'divisores(n)' para obtener los divisores de un número
def divisores(n):

    print('Los divisores de ', n, ' son:')
    # Crea bucle for para iterar sobre el rango más uno del número ingresado
    for i in range(1, n + 1):
        
        # Crea bloque condicional que evalúa si el resultado de la operación resto, entre el numero ingresado
        # y el valor de la iteración es igual a cero 
        if n % i == 0:
            print(i)
            
        else:
            continue

    print('*' * 50)

divisores(45)
divisores(31)
divisores(274)
divisores(25)
divisores(79)
divisores(54)

```

**4.2.- Crear tres funciones, cada una debe recibir un parámetro: una
lista de números.**
- La primera debe calcular el promedio de los números.


```python
# Define función que calcula promedios de una lista
def promedios(lista):

    sumatoria = 0

    for i in lista:
        sumatoria += i

    total = sumatoria / len(lista)
    print('El promedio de los números que conforman la lista es ',total)


    print('_' * 70)
    print('')


lista_uno = [1, 5, 8, 2, 9]
lista1 = [5, 8, 6, 10, 85, 79, 84]
lista2 = [25, 47, 53, 1054, 2017, 6, 2, 9, 14, 10]
lista3 = [15, 58, 236, 5478, 159647, 32, 5, 1, 1, 25, 6]

promedios(lista_uno)
promedios(lista1)
promedios(lista2)
promedios(lista3)

```

- La segunda la suma de los números.

```python
# Define función que calcula sumatoria de una lista
def suma_total(lista):

    sumatoria = 0

    for i in lista:
        sumatoria += i
    print('La suma total de los elementos que conforman la lista es ', sumatoria)


    print('_' * 70)
    print('')

suma_total(lista_uno)
suma_total(lista1)
suma_total(lista2)
suma_total(lista3)

```

**5.- Crear funciones para convertir grados entre Celsius, Fahrenheit 
y kelvin. El usuario debe ingresar un número y la unidad d 
entrada(“C”,”F”,”K”) y la de salida. Se debe elegir la función correc a
para la conveosi**n.

```python
# Crea función que transforma grados celsius a otra escala
def celsius(temperatura, salida):
    
    if salida == 'F':
        fahren = (temperatura * 9/5) + 32
        print(temperatura, 'ºC son ', fahren,'º F')
    elif salida == 'K':
        kel = temperatura + 273.15
        print(temperatura, 'ºC son ', kel,'º K')
    
# Crea función que transforma grados fahrenheit a otra escala
def fahrenheit(temperatura, salida):

    if salida == 'C':
        cels = (temperatura - 32) * 5/9
        print(temperatura, 'ºF son ', cels,'º C')
    elif salida == 'K':
        kelv = (temperatura + 459.67) * 5/9
        print(temperatura, 'ºF son ', kelv,'º K')
    
# Crea función que transforma grados kelvin a otra escala
def kelvin(temperatura, salida):

    if salida == 'C':
        cels = (temperatura - 273.15)
        print(temperatura, 'ºK son ', cels,'º C')
    elif salida == 'F':
        fahren = (temperatura * 9/5) - 459.67
        print(temperatura, 'ºK son ', fahren,'º F')
    
    
# Crea cuerpo del programa que solicita los datos para realizar la transformación de temperatura
print('Bienvenido.  Este programa convierte temperaturas entre distintas escalas')
print('-' * 65)

# Código para pedir datos de escala de entrada
print('Ingrese la escala original de temperatura')
escala_original = input('Digite: C para Celsius; F para Fahrenheit; K para Kelvin: ').upper()

# Crea bucle while que evalúa que se haya ingresado correctamente la escala de entrada
while escala_original != 'C' and escala_original != 'F' and escala_original != 'K':

    print('No se ha ingresado una escala válida.  Vuelva a intentar por favor')
    escala_original = input('Digite: C para Celsius; F para Fahrenheit; K para Kelvin: ').upper()

# Crea bloque condicional que despliega en pantalla la escala de temperatura de entrada escogida
if escala_original == 'C':
    print('La escala de entrada es Celsius')
elif escala_original == 'F':
    print('La escala de entrada es Fahrenheit')
else:
    print('La escala de entrada es Kelvin')

print('-' * 65)

# Código para ingresar valor de temperatura
temperatura = float(input('Digite la temperatura a convertir (Use el punto para separar decimales).: '))

# Crea bucle while que evalúa el mínimo de temperatura a ingresar según cada escala de temperatura
while True:
    if escala_original == 'C' and temperatura < -273.15:
        print('La temperatura mínima en Celsius es -273.15°C. No es posible realizar el cálculo.')
        temperatura = float(input('Digite la temperatura a convertir (Use el punto para separar decimales): '))
    elif escala_original == 'F' and temperatura < -459.67:
        print('La temperatura mínima en Fahrenheit es -459.67°F. No es posible realizar el cálculo.')
        temperatura = float(input('Digite la temperatura a convertir (Use el punto para separar decimales): '))
    elif escala_original == 'K' and temperatura < 0:
        print('La temperatura mínima en Kelvin es 0K. No es posible realizar el cálculo.')
        temperatura = float(input('Digite la temperatura a convertir (Use el punto para separar decimales): '))
    else:
        print('La temperatura ingresada es:', temperatura)
        break

print('-' * 65)

# Código para pedir datos de escala de salida
print('Ingrese la escala de salida de temperatura')
escala_a_convertir = input('Digite: C para Celsius; F para Fahrenheit; K para Kelvin: ').upper()

# Crea bucle while que evalúa que no se repita la escala de entrada con la de salida o que no se haya realizado el ingreso correcto
while escala_original == escala_a_convertir or escala_a_convertir != 'C' and escala_a_convertir != 'F' and escala_a_convertir != 'K':

    print('Ha ingresado la misma escala de temperatura para la entrada y salida o no ha ingresado una escala válida.  Vuelva a intentar por favor')
    escala_a_convertir = input('Digite: C para Celsius; F para Fahrenheit; K para Kelvin: ').upper()

# Crea bloque condicional que despliega en pantalla la escala de salida
if escala_a_convertir == 'C':
    print('La escala de salida es Celsius')
elif escala_a_convertir == 'F':
    print('La escala de salida es Fahrenheit')
else:
    print('La escala de salida es Kelvin')

print('-' * 65)

# Crea bloque condicional que evalúa los datos para usar la función correcta para la transformación de temperatura
if escala_original == 'C':
    celsius(temperatura, escala_a_convertir)
elif escala_original == 'F':
    fahrenheit(temperatura, escala_a_convertir)
elif escala_original == 'K':
    kelvin(temperatura, escala_a_convertir)

print('-' * 65)
print('El programa ha finalizado')
```
