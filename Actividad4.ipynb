1. Escriba un programa en Python para crear una clase que represente un círculo.  Incluya métodos para calcular su área y su perimetro.

```python
# Importa el módulo math
import math

# Crea función Circulo para el cálculo de áreas y perímetros de un circulo según valor de su radio
class Circulo:

    # Crea constructor de clase con atributo radio
    def __init__(self, radio):
        self.radio = radio

    # Crea método para el cálculo del área de un circulo según valor del radio ingresado por el usuario
    def area(self):
        return math.pi * self.radio ** 2

    # Crea método para el cálculo del perímetro de un circulo según valor del radio ingresado por el usuario
    def perimetro(self):
        return 2 * math.pi * self.radio


# Crea función para solicitar al usuario el radio del círculo y verificar la entrada
def solicitar_radio():

    # Crea bucle while que contiene bloque para manejo de excepciones en el ingreso de la información
    while True:

        # Bloque de manejo de excepciones
        try:
            radio = float(input('Ingrese el radio del círculo: '))

            # Crea bloque condicional que evalúa que los datos ingresados satisfagan el correcto cálculo de los métodos de clase
            if radio <= 0:

                # Levanta excepción 'ValueError' en caso de ingreso erróneo
                raise ValueError('El radio debe ser un número positivo mayor que cero.')

            
            return radio

        # Toma el lanzamiento de la excepción e imprime mensaje de advertencia
        except ValueError as e:
            print('Error:', e)

# Crea bucle while para solicitar el ingreso del valor del radio del círculo
while True:

    # Crea variable radio que invoca la fucnión solicitar_radio() para obtener el valor del radio
    radio = solicitar_radio()

    # Crea variable circulo que invoca la clase Circulo con argumento atributo variable radio
    circulo = Circulo(radio)

    print('´´' * 30)

    # Imprime los resultados esperados
    print('El área de un círculo de radio', radio, 'es:', circulo.area())
    print('El perímetro de un círculo de radio', radio, 'es:', circulo.perimetro())

    # Crea variable continuar para consultar si el usuario desea realizar otro cálculo
    continuar = input('¿Desea calcular el área y el perímetro de otro círculo? (s/n): ')
    print('´´' * 30)
    
    # Crea bloque condicional que evalúa la respuesta del usuario para seguir operando o salir del programa
    if continuar.lower() != 's':
        break

print('El programa ha finalizado')

```

2. Escriba un programa en Python para crear una clase 'persona'.  Incluya atributos tales como nombre, país y fecha de nacimiento.  Implemente un método para determinar la edad de la personaa.

```python
# Del módulo datetime importamos el método datetime
from datetime import datetime

# Crea clase persona
class Persona:

    # Crea constructor de clase con atributos: Nombre, País y Fecha de Nacimiento
    def __init__(self, nombre, pais, fecha_nacimiento):
        self.nombre = nombre
        self.pais = pais
        self.fecha_nacimiento = fecha_nacimiento

    # Crea método para el cálculo de la edad de las instancias creadas
    def calcular_edad(self):
        fecha_actual = datetime.now()
        edad = fecha_actual.year - self.fecha_nacimiento.year
        if (fecha_actual.month, fecha_actual.day) < (self.fecha_nacimiento.month, self.fecha_nacimiento.day):
            edad -= 1
        return edad

# Crea función para solicitar al usuario una fecha de nacimiento y verifica la entrada
def solicitar_fecha_nacimiento():

    # Crea bucle while que evaluará si el ingreso de la fecha manejando excepciones
    while True:

        # Crea variable fecha_str para solicitar el ingreso de la fecha de nacimiento
        fecha_str = input('Ingrese la fecha de nacimiento de la persona (en formato dd/mm/aaaa): ')

        # Bloque de manejo de excepciones
        try:
            
            fecha_nacimiento = datetime.strptime(fecha_str, '%d/%m/%Y')
            
            # Crea bloque condicional que evalúa si la fecha de nacimiento es válida
            if fecha_nacimiento > datetime.now():

                # Levanta ValueError si la fecha ingresada es posterior a la fecha actual
                raise ValueError('La fecha de nacimiento no puede ser posterior a la fecha actual.')

            # Devuelve fecha de nacimiento en caso de que se haya ingresado una fecha válida
            return fecha_nacimiento

        # Permite que el programa siga corriendo en caso de que se produzca un error
        except ValueError as e:
            print('Error:', e)

# Crea bucle while True con manejo de excepciones y reingreso de datos incorrectos
while True:

    # Crea bloque de manejo de excepciones
    try:

        # Crea variables para el ingreso de los atributos de la instancias a crear
        nombre = input('Ingrese el nombre de la persona: ')
        pais = input('Ingrese el país de la persona: ')
        fecha_nacimiento = solicitar_fecha_nacimiento()

        # Crear una instancia de la clase Persona con las variables ya creadas por solicitud input()
        persona = Persona(nombre, pais, fecha_nacimiento)

        # Imprime el nombre de la instancia creada y su edad calculada
        print('La edad de', persona.nombre, 'es', persona.calcular_edad(), 'años.')

        # Consulta al usuario si quiere seguir usando el programa
        continuar = input('¿Desea ingresar otra persona? (s/n): ')
        print('´´' * 30)
        
        # Bloque condicional que evalua la respuesta de variable 'continuar'
        if continuar.lower() != 's':
            
            # Instrucción para salir del bucle si la respuesta es distinta a 's'
            break  

    
    except Exception as e:
        print('Ha ocurrido un error:', e)

print('´´' * 30)
print('El programa ha finalizado')

```

3. Escriba un programa en Python para crear una clase 'Calculadora'.  Incluya métodos de operaciones aritméticas básicas: suma, resta, multilpicación y divisiónn.

```python
# Crea clase calculadora
class Calculadora:

    # Crea métodos para las operaciones aritméticas básicas
    def suma(self, a, b):
        return a + b

    def resta(self, a, b):
        return a - b

    def multiplicacion(self, a, b):
        return a * b

    def division(self, a, b):
        if b == 0:
            raise ZeroDivisionError("No se puede dividir entre cero.")
        return a / b

# Crea una instancia de la clase Calculadora
calculadora = Calculadora()

# Crea diccionario de operaciones disponibles
operaciones = {
    '1': 'suma',
    '2': 'resta',
    '3': 'multiplicacion',
    '4': 'division'
}

# Crea bucle while para asegurar que se indiquen datos correctos
while True:
    print('Operaciones disponibles:')
    print('1. Suma')
    print('2. Resta')
    print('3. Multiplicación')
    print('4. División')

    # Crea variable opcion que permite elegir la operación deseada
    opcion = input('Seleccione una operación (1/2/3/4): ')

    # Crea bloque condicional para verificar que se haya ingresado una opción válida
    if opcion in operaciones:

        # Bloque para manejo de excepciones
        try:
            num1 = float(input('Ingrese el primer número: '))
            num2 = float(input('Ingrese el segundo número: '))

            # Crea variable para obtener el nombre del método de la calculadora correspondiente a la operación seleccionada
            nombre_metodo = operaciones[opcion]
            
            # Crea variable resultado que realiza la operación correspondiente
            resultado = getattr(calculadora, nombre_metodo)(num1, num2)

            # Muestra al usuario el resultado de la operación
            print('Resultado de la', nombre_metodo.capitalize() + ' es:', resultado)

            # Instrucción para salir del bucle después de procesar una operación válida
            break

        # Captura las excepciones que levantadas por el programa
        except ValueError:
            print('Por favor, ingrese números válidos.')
        except ZeroDivisionError as e:
            print('Error:', e)
            print('Por favor, ingrese un divisor diferente de cero.')

        print('´´' * 30)
    # Informa que la opción elegida no es válida y vuelve al inicio del bucle
    else:
        print('Opción no válida. Por favor, seleccione una opción válida.')
        print('´´' * 30)

print('´´' * 30)
print('El programa ha finalizado')
```

5. Escriba un programa en Python para crear una clase que represente un carro de compras. Incluya métodos para agregar y quitar ítems y para calcular el precio total.l.

```python
# Crea clase CarroDeCompraas
class CarroDeCompras:

    # Crea constructo de clase
    def __init__(self):

        # Crea estructura de datos diccionario para almacenar los items agregados
        self.items = {}

    # Crea método para agregar items al carro estableciendo la cantidad como una unidad si es que el usuario no indica ese atributo
    def agregar_item(self, nombre, precio, cantidad=1):

        # Crea bloque de manejo de excepciones
        try:
            precio = float(precio)
            cantidad = int(cantidad)

            # Crea bloque condicional que evalúa los valores de precio y cantidad para evitar que se ingresen valores menores o iguales a cero
            if precio <= 0 or cantidad <= 0:

                # Levanta excepcion ValueError
                raise ValueError("El precio y la cantidad deben ser valores positivos.")

            #  Bloque condicional que evalúa pre-existencia del producto en el carro y agregar la nueva cantidad
            if nombre in self.items:
                self.items[nombre]['cantidad'] += cantidad

            # Si el producto no está previamente en el diccionario ejecuta linea que agrega nueva clave:valor con precio y cantidad
            else:
                self.items[nombre] = {'precio': precio, 'cantidad': cantidad}

        # toma las excepciones levantadas e imprime mensaje de error informativo
        except ValueError as e:
            print(f"Error al agregar el ítem '{nombre}': {e}")

    # Crea método para quitar items del carro estableciendo la cantidad como una unidad si es que el usuario no indica ese atributo 
    def quitar_item(self, nombre, cantidad=1):

        # Crea bloque de manejo de excepciones
        try:
            cantidad = int(cantidad)

            # Crea bloque condicional que evalúa la cantidad para evitar que se ingresen valores menores o iguales a cero
            if cantidad <= 0:

                # Levanta excepcion ValueError
                raise ValueError("La cantidad debe ser un valor positivo.")

            # Crea bloque condicional que evalua la pre-existencia del producto en el carro
            if nombre in self.items:

                # Si el producto ya existe en el carro resta la cantidad establecida o la indicada por el usuario
                self.items[nombre]['cantidad'] -= cantidad

                # Bloque condicional que evalúa la cantidad de un producto en el carro, eliminandolo si es menor o igual a cero
                if self.items[nombre]['cantidad'] <= 0:
                    del self.items[nombre]

        # toma las excepciones levantadas e imprime mensaje de error informativo
        except ValueError as e:
            print(f"Error al quitar el ítem '{nombre}': {e}")

    # Crea método para calcular el precio total de los productos incluidos en el carro de compras
    def calcular_precio_total(self):

        # Crea bloque de manejo de excepciones
        try:
            total = sum(item['precio'] * item['cantidad'] for item in self.items.values())
            return total
        except Exception as e:
            print(f"Error al calcular el precio total: {e}")

```

```python
# Líneas de código de ejemplo para ejecutar el programa

# Crea ejemplar de clase CarroDeCompras()
carro = CarroDeCompras()

# Crea líneas de código con método agregar_item() para ingresar productos al carro
carro.agregar_item('Leche', 2.5, 2)
carro.agregar_item('Pan', 1.0)
carro.agregar_item('Manzanas', 0.5, 6)

# Usa método de clase para calcular costo total de los productos incluidos en el carro
precio_total = carro.calcular_precio_total()

# Bloque condicional que evalúa si el monto del carro es mayor a cero e imprime total en pantalla
if precio_total is not None:
    print("Precio total del carro de compras:", precio_total)
```
